This generates a list comprising uppercase alphabets that correspond to the hidden columns of an Excel worksheet. 
- Need to somehow convert the hidden_cols list into a list of names of the hidden columns. 
- Use Python’s built-in library, string, and its 
    - `.ascii_uppercase` attribute

```python
# List of indices corresponding to all hidden columns
hidden_cols_idx = [string.ascii_uppercase.index(col_name) for col_name in hidden_cols]

# Find names of columns corresponding to hidden column indices
hidden_cols_name = df.columns[hidden_cols_idx].tolist()
print(hidden_cols_name)
```

```python
# Import modules
import openpyxl 
import pandas as pd
import string

# Read Excel file as Pandas DataFrame
df = pd.read_excel("data/titanic.xlsx")

# Open an Excel workbook
workbook = openpyxl.load_workbook("data/titanic.xlsx")

# Create a `Worksheet` object 
worksheet = workbook[workbook.sheetnames[0]]

# List of indices corresponding to all hidden rows
hidden_rows_idx = [
    row - 2
    for row, dimension in worksheet.row_dimensions.items() 
    if dimension.hidden
]

# List of indices corresponding to all hidden columns
hidden_cols_idx = [
    string.ascii_uppercase.index(col_name) 
    for col_name in [
        col 
        for col, dimension in worksheet.column_dimensions.items() 
        if dimension.hidden
    ] 
]

# Find names of columns corresponding to hidden column indices
hidden_cols_name = df.columns[hidden_cols_idx].tolist()

# Drop the hidden columns and rows
df.drop(hidden_cols_name, axis=1, inplace=True)
df.drop(hidden_rows_idx, axis=0, inplace=True)

# Reset the index
df.reset_index(drop=True, inplace=True)

df
```

unfortunately

```python
from openpyxl import load_workbook

# Load the source and target workbooks
source_wb = load_workbook('output3/example.xlsx')
target_wb = load_workbook('output3/example2.xlsx')  # keep_vba is needed if the workbook has macros

# Get the sheet to copy from the source workbook
source_sheet = source_wb['NewSheet2']

# Create a new sheet in the target workbook
target_sheet = target_wb.create_sheet(title="CopiedSheet")

# Copy data from source sheet to target sheet
for row in source_sheet.iter_rows(values_only=True):
    target_sheet.append(row)

# Save the target workbook
target_wb.save('output3/example2.xlsx')
```

## Exercise 7 - where function with Pandas
- The values that fit the condition remain the same. 
- The other values are updated

```python
items = pd.DataFrame({"item": ["apple", "banana", "orange", "milk", "egg", "cheese"],
                      "category": ["fruit", "fruit", "fruit", "dairy", "dairy", "dairy"],
                      "price": [3.5, 2.9, 2.8, 4.5, 6.5, 8.0]})

items
```

```python
items["new_price"] = items["price"].where(items["price"] > 3, items["price"] * 1.1)

items

# The values that fit the condition (price > 3) remain the same. The other values are updated.
```

```python
items["price_increased"] = 0

items["price_increased"] = items["price_increased"].where( items["new_price"] - items["price"] <= 0, 1)
```

```python
items
```

## Exercise 8 - NumPy where function

* The where function assigns a value based on one set of conditions. 

Parameter/ arguments:
- arg 1: Condition
- arg 2: Satisfied expression (True)
- arg 3: Unsatisfied expression (False)

```python
# Updated both the values that fit and do not fit the condition.
items["new_price_npwhere"] = np.where(items["price"] > 3, items["price"] * 0.95, items["price"] * 1.1)

items.drop('new_price_2', axis = 1, inplace= True)
```

## Exercise 9 - NumPy select function

* The select function accepts multiple sets of conditions and 
- assign a different value for each set of conditions.

```python
conditions = [(items["price"] < 3) & (items["category"] == "fruit"),
              (items["price"] < 5) & (items["category"] == "dairy")]

values = [items["price"] * 1.1, items["price"] * 1.05]

items["new_price_npselect"] = np.select(conditions, values, default = items["price"])

items
```

## Exercise 17 - qcut function

* With the cut function, we do not have any control over how many values fall into each bin.
* We can only specify the bin edges.

* This is where we need to learn about the qcut function. 
* It can be used to divide the values into buckets in a way that each bucket contains approximately the same number of values.

* We have 4 buckets and each one contains almost the same number of values.
* In case of 4, the buckets are also called quartiles. 
    - One quarter of the total number of values are in the first quartile, 
    - one half are in the first two buckets, 
    - and so on.

```python
df_cutit["col_a_binned_qcut"] = pd.qcut(df_cutit["col_a"], q=4)

df_cutit.head()
```

```python
df_cutit["col_a_binned_qcut"].value_counts()
```

* The qcut function allows for customizing the bucket sizes. 
* Let’s create 3 buckets. The first one contains the smallest 50 percent of values (i.e. lower half). Then, we divide the upper half into two bins.

```python
pd.qcut(df_cutit.col_a, q=[0, .50, .75, 1]).value_counts()
```

### find elements in a pandas Series that contain the substring medium.com within a URL

For matching strings with special characters

.str.match() uses regular expressions (regex) for pattern matching. 
- leverage the full power of regex, including handling special characters.

.str.match() method is used to check if each string in a Series matches a given regular expression pattern. 
- It returns a boolean Series indicating whether each element in the Series matches the pattern.

This method is a bit like a strict bouncer at a club - it only lets in strings that match a certain pattern exactly.

.str.match() is case sensitive and matches the pattern exactly.

However, since .str.match() checks if the whole string matches the pattern from the beginning, it might be more suitable to use .str.contains() for this specific use case.

### Handling Special Characters

- Dot (.): Matches any character except a newline. To match a literal dot, use \..
- Asterisk (*): Matches zero or more of the preceding element. To match a literal asterisk, use \*.
- Plus (+): Matches one or more of the preceding element. To match a literal plus, use \+.
- Question mark (?): Matches zero or one of the preceding element. To match a literal question mark, use \?.
- Parentheses (()): Used for grouping. To match literal parentheses, use \( and \).
- Brackets ([]): Used for character classes. To match literal brackets, use \[ and \].
- Caret (^): Matches the start of a string. To match a literal caret, use \^.
- Dollar sign ($): Matches the end of a string. To match a literal dollar sign, use \$


```python
emails = pd.Series(['bob@gmail.com', 'alice@yahoo.com', 'charlie@gmail.com', 'david@outlook.com'])
gmail_emails = emails[emails.str.match('.*@gmail.com$')]
print(gmail_emails)
```

# Multi-indexing
- what a multiindex is
    - An index is a column in a DataFrame that ‘uniquely’ identifies each row.
    - A multiindex is when there is more than one index. Other names are 
        - multiple index and 
        - hierarchical index.
    - Multiindex can also refer to multiple header levels
- where and when to use multindexing, 
- functions that generate a multiindex, and 
- how to collapse it into a single index.

**Advantages of multiple indices**

- For holding data that is higher-dimensional and contains hierarchies or levels.
- Provides a helpful visual display of the hierarchy levels in a table format.
- It allows for efficient selection and manipulation of hierarchical data using functions such as 
    - df.xs() and 
    - df.unstack().

**Disadvantages of multiindex**

- The format does not allow for direct plotting of graphs.
- There are no performance benefits when using the multiindex.

### Pandas Styler Object
Any Pandas DataFrame contains a Styler object which can be accessed by calling `.style`
- This returns a pandas.io.formats.style.Styler object

The precision of the displayed values has changed
- behind the scenes the Styler object is converting the table into HTML and applying certain styles to it
- call `.style.to_hmtl()` on your DataFrame to see the HTML printed as a string.

The Styler treats the underlying data value different from the display value — the latter is a string representation of the former. 
- When we format the values using the Styler object and its methods, 
    - only the display value gets altered (using CSS) so the underlying DataFrame remains untouched.
    
###### Explaination
- data.style: Creates a Styler object that uses CSS to customize various components of a Dataframe.

- data.style.applymap(): All these components are passing to the Styler object from a style function using applymap(). 
    - background-color: white;
    - font-size: red;

## Exercise 7.6.1: Highlighting rows based on categorical values

- First, the style function, highlight_rows(), now takes in each row as an argument, as opposed to the previous highlight_cells() function which takes in each cell value as an argument.

- Second, since we are applying a style function row-wise, we use .apply() with axis=1 instead of .applymap().

- Third, because we are applying the function row-wise, the output of the style function is a list of CSS strings, instead of a single string.

```python
def highlight_rows(row):
    value = row.loc['species']
    if value == 'versicolor':
        color = '#FFB3BA' # Red
    elif value == 'setosa':
        color = '#BAFFC9' # Green
    else:
        color = '#BAE1FF' # Blue
    return ['background-color: {}'.format(color) for r in row]
```

```python
df.style.apply(highlight_rows, axis=1)
```

## Exercise 7.7: Conditional text formatting

CSS attribute-value pair for cell highlighting that we have configured takes the following format: 
- `background-color: <set color>`. 
    
We can apply other types of conditional formatting by specifying different CSS attribute-value pair formats.

Customize the following:
- font color: 'color: `<color/hex code>`'
- font type: 'font-family: `<font_type>`'
- font size: 'font-size: `<font size in px>`'
- font weight: 'font-weight: `<font weight>`'
    
Specify more than one type of attribute in a single style function by chaining them with a semicolon, like this: 
- `background-color: 'yellow'; color: 'blue'; font-size: '15px'`

```python
def highlight_cells(value, color_true, color_false, criteria):
    if value == criteria:
        color = color_true 
    else:
        ''
    return 'color: {}'.format(color)
```

```python
data.style.applymap(highlight_cells, color_true = 'green', color_false = '', criteria = 2.55)
```

## Example:  Set font color and weight when sepal length or width is between 3.5mm and 5.5mm

Style function should return needs to be in the 'color: <set color>; font-weight: <set font weight>’ format.

```python
def format_sepal(val):
    condition = (val >= 3.5) & (val <= 5.5)
    font_color = 'red' if condition else 'black'
    font_weight = 'bold' if condition else 'normal'
    return 'color: {}; font-weight: {}'.format(font_color, font_weight)

df.style.apply(highlight_rows, axis=1)\
        .applymap(format_sepal, subset=['sepal_length', 'sepal_width'])
```

## Example: Set font size and type when petal length or width is between 1.5mm and 3.5mm

Define another style function to return a CSS string in the 'font-size: <set font size>; font-family: <set font type>’ format.

```python
def format_petal(val):
    condition = (val >= 1.5) & (val <= 3.5)
    font_size = '15px' if condition else '1em'
    font_type = 'cursive' if condition else 'serif'
    return 'font-size: {}; font-family: {}'.format(font_size, font_type)

df.style.apply(highlight_rows, axis=1).applymap(format_sepal, subset=['sepal_length', 'sepal_width'])\
        .applymap(format_petal, subset=['petal_length', 'petal_width'])
```

```python
def min_max(col):
    # empty property list to be filled in the loop
    props = ['font-weight: bold;'] * len(col)
    for idx, value in enumerate(col):
        # color max value green
        if value == max(col):
            props[idx] += 'background-color: lightgreen;'
        # color min value red
        if value == min(col):
            props[idx] += 'background-color: lightcoral'
        
    return props
df.style.apply(min_max)
```

# 15.4. Outliers Detection

<a id='outlier'></a>
[Back to table of contents](#toc)

```python
# Outliers Detection
def detect_outliers(data, threshold=3):
    z_scores = (data - data.mean()) / data.std()
    outliers = np.abs(z_scores) > threshold
    return outliers

outliers = detect_outliers(df['A'])
df['A'][outliers] = np.nan  # Replace outliers with NaN
```

# Pivot tables in Pandas and Handling Multi-Index Data 

Pandas pivot_table function and how to use the various parameters it offers

Pivot table syntax says that:
- for every ‘index’, 
- return the ‘aggfunc’ of the ‘values’ column(s), 
- segregated (further grouped) by ‘columns’.

**Pivot_table() syntax**

`pandas.pivot_table(data, 
                  values=None, 
                  index=None, 
                  columns=None, 
                  aggfunc='mean', 
                  fill_value=None, 
                  margins=False, 
                  dropna=True, 
                  margins_name='All', 
                  observed=False, 
                  sort=True)`
                

**Required parameters:**
- data: the data source, in the form of pandas DataFrame.

**One of the two parameters below must be present to group the data:**
- index: column(s) to group the data row-wise.
- columns: column(s) to group the data column-wise.

**Optional parameters:**
- values: column(s) to aggregate using the aggfunc function.
- aggfunc: Function used to aggregate the values by.
- fill_value: value to replace missing values with.
- dropna: whether to remove entire rows or columns that contain only NaN values.
- margins: whether to include row and column subtotals.
- margins_name: label names for the row and column subtotals.
- observed: displays only observed values for categorical groupers.
- sort: whether to sort the resulting row indices.

**Returns:**

A DataFrame with the summarized data.

[Back to table of contents](#toc)

how : {'left', 'right', 'outer', 'inner', 'cross'}, default 'inner'
    Type of merge to be performed.

* **left**: use only keys from left frame, similar to a SQL left outer join;
  preserve key order.
* **right**: use only keys from right frame, similar to a SQL right outer join;
  preserve key order.
* **outer**: use union of keys from both frames, similar to a SQL full outer
  join; sort keys lexicographically.
* **inner**: use intersection of keys from both frames, similar to a SQL inner
  join; preserve the order of the left keys.
* **cross**: creates the cartesian product from both frames, preserves the order
  of the left keys.

# ---------------------------- Merge "not matching" time-series data

* Time-series data might include measurements taken at very short time periods (e.g. at the level of seconds).
* Therefore, when we merge two DataFrames consist of time series data, we may encounter measurements off by a second or two. 
* For such cases, Pandas provides a “smart” way of merging done by `merge_asof`.
* Assume we are merging DataFrames A and B. 
    - If a row in the left DataFrame does not have a matching row in the right DataFrame, 
    - `merge_asof` 
        - allows for taking a row whose value is close to the value in left DataFrame.

## Exercise 15 - merge_asof

For each row in the left DataFrame:

* A “backward” search selects the last row in the right DataFrame whose ‘on’ key is less than or equal to the left’s key.

* A “forward” search selects the first row in the right DataFrame whose ‘on’ key is greater than or equal to the left’s key.

* A “nearest” search selects the row in the right DataFrame whose ‘on’ key is closest in absolute distance to the left’s key.

#### Rank the output using this function 

```python
def get_rankings(item, df = fruits_expanded_v1):
    
    # Empty dict for results
    item_count_dict = {}
    
    # For every tag in df
    for i in range(df.shape[1]):
        
        # Calculate % of cases that tagged the item
        val_counts = df.iloc[:,i].value_counts(normalize = True)
        if item in val_counts.index:
            item_counts = val_counts[item]
        else:
            item_counts = 0
        
        # Add score to dict
        item_count_dict["tag_{}".format(i)] = item_counts
    
    return item_count_dict
```

```python
def boolean_df(item_lists, unique_items):

    # Create empty dict
    bool_dict = {}
    
    # Loop through all the tags
    for i, item in enumerate(unique_items):
        
        # Apply boolean mask that returns a True-False list of whether a tag is in a taglist
        bool_dict[item] = item_lists.apply(lambda x: item in x)
            
    # Return the results as a dataframe
    return pd.DataFrame(bool_dict)
```

## Exercise 13 - groupby and rank() function

* The rank function is used for assigning a rank to the rows based on the values in the given column. 
* We can use the rank and the groupby functions to rank rows within each group separately.

method parameter defines how to rank the group of records that have the same value (i.e. ties):

* average: average rank of the group (default values)

* min: lowest rank in the group

* max: highest rank in the group

* first: ranks assigned in order they appear in the array

* dense: like ‘min’, but rank always increases by 1 between groups.

```python
sales["rank"] = sales.groupby("store")["price"].rank(ascending=False, method="dense")
                              
sales.head()
```

## Exercise 15 - groupby and expending() function

```python
!pip install textblob
```

```python
import pandas as pd
from textblob import TextBlob

def remove_white_space(df: pd.DataFrame):
    df['text'] = df['text'].apply(lambda row: row.strip())
    return df

def get_sentiment(df: pd.DataFrame):
    df['sentiment'] = df['text'].apply(lambda row: TextBlob(row).sentiment[0])
    return df
```

```python
df = pd.DataFrame({'text': ["It is a beautiful day today  ", "  This movie is terrible", '  I hate it here ']})
```

```python
df = (df.pipe(remove_white_space).pipe(get_sentiment))
print(df)
```

# Comparing Datasets to check Changes

compares data from the current quarter with data from the previous quarter, saves or appends any changed rows and new entries, you'll need to follow these steps:

- Load Data: Retrieve data from both the current and previous quarters.
- Compare Data: Identify changes and new entries.
- Save Results: Save the changes and new entries to a file or database.

```python
import pandas as pd

def load_data(file_path):
    """Load CSV data into a DataFrame."""
    return pd.read_csv(file_path)

def compare_data(current_df, previous_df):
    """Compare current and previous DataFrames to find new entries and changes."""
    # Find new entries: rows present in current_df but not in previous_df
    new_entries = current_df[~current_df.apply(tuple, 1).isin(previous_df.apply(tuple, 1))]
    
    # Find changed entries: rows that exist in both but have different values
    common_entries = pd.merge(current_df, previous_df, on='ID', suffixes=('_current', '_previous'))
    changes = common_entries[
        (common_entries.filter(like='_current') != common_entries.filter(like='_previous')).any(axis=1)
    ]
    
    # We will need to select only the columns from the current DataFrame
    changes = changes[current_df.columns]
    
    return new_entries, changes

def save_results(new_entries, changes, file_path):
    """Save new entries and changes to a CSV file."""
    # Combine new entries and changes
    results = pd.concat([new_entries, changes]).drop_duplicates()
    results.to_csv(file_path, index=False)

def main():
    # File paths
    current_quarter_file = 'current_quarter.csv'
    previous_quarter_file = 'previous_quarter.csv'
    results_file = 'changes_and_new_entries.csv'
    
    # Load data
    current_df = load_data(current_quarter_file)
    previous_df = load_data(previous_quarter_file)
    
    # Compare data
    new_entries, changes = compare_data(current_df, previous_df)
    
    # Save results
    save_results(new_entries, changes, results_file)
    
    print(f"New entries and changes have been saved to {results_file}")

if __name__ == "__main__":
    main()
```
